```python
import numpy
import pylab
import finess.util
import finess.params.util
import finess.dim2
import generate_iniparams
```

```python
params = finess.params.util.read_params('output2/parameters2.ini.dump', generate_iniparams.parameter_list)
```

```python
xlow = params['grid', 'xlow']
xhigh = params['grid', 'xhigh']
ylow = params['grid', 'ylow']
yhigh = params['grid', 'yhigh']
mx = params['grid', 'mx']
my = params['grid', 'my']
```

```python
dx = (xhigh - xlow) / float(mx)
dy = (yhigh - ylow) / float(my)
```

```python
X = numpy.linspace(xlow + 0.5*dx, xhigh - 0.5*dx, num=mx, endpoint = True)
```

```python
%matplotlib inline
```

```python
for frame_idx in range(100 + 1):
    t, q, a = finess.dim2.read_qa(params, frame_idx)
    pylab.plot(X, q[:, 1 - 1, 6 - 1])
    pylab.title(str(frame_idx))
    pylab.show()
```
