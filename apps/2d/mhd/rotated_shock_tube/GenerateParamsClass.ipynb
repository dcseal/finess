```python
class_name = "Params"
global_variable_name = "params"
```

```python
header_guard_macro_name = "_" + class_name.upper() + "_H_"
header_guard_macro_name
```

```python
header_filename = class_name + ".h"
cpp_filename = class_name + ".cpp"
```

```python
import inspect, os
```

```python
current_python_filename = inspect.getfile(inspect.currentframe())
```

```python
header_comments_file = """/// @file %(header_filename)s
/// Generated by %(current_python_filename)s
""" % \
{"header_filename": header_filename, "current_python_filename": current_python_filename}
```

```python
header_comments_usage = """
// This file, %(header_filename)s, and %(cpp_filename)s, together define a class %(class_name)s,
// and a global variable %(global_variable_name)s.
//
// To use %(global_variable_name)s,
// 1. Add #include "%(header_filename)s" to all the files that use %(global_variable_name)s;
// 2. Add %(global_variable_name)s.init(<parameter_filename>); to main function, where
//    <parameter_filename> is to be replaced by the actual parameter filename;
// 3. Make sure -I. is in CXXFLAGS of Makefile;
// 4. Make sure %(cpp_filename)s is compiled and linked.
""" % \
{"header_filename": header_filename,
 "cpp_filename": cpp_filename,
 "class_name": class_name,
 "global_variable_name": global_variable_name}
```

```python
print header_comments_usage
```

```python
header_head = """%(header_comments_file)s

#ifndef %(header_guard_macro_name)s
#define %(header_guard_macro_name)s

#include <string>
#include "dog_ini.h"
#include "util.h"

class %(class_name)s;
extern %(class_name)s %(global_variable_name)s;

""" % \
{"header_comments_file": header_comments_file,
 "header_guard_macro_name": header_guard_macro_name,
 "class_name": class_name,
 "global_variable_name": global_variable_name
 }
```

```python
print header_head
```

```python
header_tail = """
#endif
"""
print header_tail
```

```python
class_declaration_contents = """
public:
    struct ReconstructionMethod{
        enum enum_type {A, B, C, DEFAULT};
    };
private:    
    ReconstructionMethod::enum_type reconstruction_method;
public:
    inline ReconstructionMethod::enum_type get_reconstruction_method(){
        return this->reconstruction_method;
    }
private:
    int meqn;
public:
    inline int get_meqn(){
    return this->meqn;
    }
"""
```

```python
class_declaration = """
class Params{
public:
    Params(){}
    void init(const std::string& inputFilename);

%(class_declaration_contents)s    
};
""" % \
{"class_declaration_contents": class_declaration_contents}
```

```python
print class_declaration
```

```python
cpp_comments_file = """/// @file %(cpp_filename)s
/// Generated by %(current_python_filename)s
""" % \
{"cpp_filename": cpp_filename, "current_python_filename": current_python_filename} 
```

```python
print cpp_comments_file
```

```python
cpp_head = """%(cpp_comments_file)s
#include "%(header_filename)s"

%(class_name)s %(global_variable_name)s;

""" % \
{"cpp_comments_file": cpp_comments_file,
 "header_filename": header_filename,
 "class_name": class_name,
 "global_variable_name": global_variable_name}
```

```python
print cpp_head
```

```python
cpp_init_method_contents = """
    string reconstruction_method_str = ini_doc["reconstruction"]["method"];
    if(reconstruction_method_str == "")
    terminate("reconstruction.method is missing.");

    {
        this->reconstruction_method =
        reconstruction_method_str == "A" ? Params::ReconstructionMethod::A :
        reconstruction_method_str == "B" ? Params::ReconstructionMethod::B :
        reconstruction_method_str == "C" ? Params::ReconstructionMethod::C :
        Params::ReconstructionMethod::DEFAULT;
    }

    if(this->reconstruction_method == Params::ReconstructionMethod::DEFAULT)
        terminate("reconstruction.method must be one of the following: A, B, C.");
    
    string meqn_str = ini_doc["dogParams"]["meqn"];
    if(meqn_str == "")
        terminate("dogParams.meqn is missing.");

    {
        this->meqn = stringToAny<int>(meqn_str);
    }

    if(this->meqn <= 0)
        terminate("dogParams.meqn is expected to be positive.");
"""
```

```python
cpp_init_method = """
void Params::init(const std::string& inputFilename){
    using std::string;
    IniDocument ini_doc;
    ini_doc.initFromFile(inputFilename);

%(cpp_init_method_contents)s

}

""" % \
{"cpp_init_method_contents": cpp_init_method_contents}
```

```python
print cpp_init_method
```

```python
header_result = header_head + class_declaration + header_tail
```

```python
print header_result
```
